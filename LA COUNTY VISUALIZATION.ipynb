```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
la = pd.read_csv("LA_County_Covid19_cases_deaths_date_table.csv")
la.head()
```

```python
la = la[["date_use", "total_cases","new_case","total_deaths","new_deaths"]]
la["date_use"] = pd.to_datetime(la["date_use"])
# la["date_use"] = la["date_use"].astype(str).astype(int)
la.head()
# la.info()
# dtm = lambda x: datetime(la["date_use"][5:7])
# months = list(map(dtm,la["date_use"]))
```

```python
ax = plt.gca()

# la_plot.set_xticklabels(la["date_use"], rotation=45, ha = "right")
# plt.xticks(la["date_use"])
# 

# ax = (la["date_use"].groupby(la["date_use"].dt.month).count()).plot(kind="bar", color='green')
# la.hist(x="date_use", y="new_case",label ="daily new cases",ax=ax)
la.plot(kind='line',x='date_use',y='total_cases',label="total cases",ax=ax, color ='blue',figsize=(10,15))
# la.plot(kind="line",x='date_use',y='total_deaths',label='cumulative deaths',ax=ax, color='red',figsize=(10,15))
ax.set_ylabel("Total_cases")
ax.set_title("Total number of cases by month")
plt.legend(loc='upper right')
plt.show()
```

```python
ax = plt.gca()
la.plot(kind='line',x='date_use',y='new_case',label='new cases',ax=ax,color = 'green', figsize=(10,15))
la.plot(kind="line",x='date_use',y='total_deaths',label='cumulative deaths',ax=ax, color='red',figsize=(10,15))
plt.legend(loc='upper right')
# ax.set_ylabel("")
ax.set_title("New cases number by month")
plt.show()
```
